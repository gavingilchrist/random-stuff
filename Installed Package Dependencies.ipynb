```python
import subprocess

env_packages = [i.split()[0] 
                for i in subprocess.run(["pip", "list"], 
                                        capture_output=True,
                                        text=True).stdout.split('\n')[2:]
                if i != '']

req_by = {i: [k.strip() 
              for k in [j for j in subprocess.run(["pip", "show", i], 
                                                  capture_output=True,
                                                  text=True).stdout.split('\n')
                        if 'Required-by' in j][0].split(':')[1].split(',')
                        if k.strip() != '']
          for i in env_packages}

def ult_req_by(pkg):
    user_pkgs = req_by[pkg]
    if len(user_pkgs) == 0:
        return []
    else:
        outlist = [*user_pkgs]
        for i in user_pkgs:
            for j in ult_req_by(i):
                if j not in outlist:
                    outlist += [j]
        return outlist

for i in env_packages:
    print(i)
    print(ult_req_by(i))
    print()
```
