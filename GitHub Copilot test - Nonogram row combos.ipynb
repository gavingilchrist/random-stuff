```python
def recursive_multiply(a, b):
    """
    Multiply two numbers using recursion.
    :param a: First number (integer)
    :param b: Second number (integer)
    :return: Product of a and b
    """
    # Base case: if one of the numbers is 0, the result is 0
    if b == 0:
        return 0
    
    # Handle negative numbers
    if b < 0:
        return -recursive_multiply(a, -b)
    
    # Recursive case
    return a + recursive_multiply(a, b - 1)

recursive_multiply(-3, -5)
```

```python
from itertools import permutations

def generate_arrangements(total_cells, groups):
    """
    Generate all possible cell arrangements based on the input groups of filled cells.

    :param total_cells: Total number of cells (int)
    :param groups: List of numbers where each number represents a group of consecutive filled cells
    :return: List of all possible cell arrangements (list of strings)
    """
    # Calculate the total number of filled cells
    total_filled = sum(groups)
    # Calculate the total number of empty cells
    total_empty = total_cells - total_filled

    # There must be at least one empty cell between groups, so calculate separators
    separators = len(groups) - 1

    # If there aren't enough empty cells to separate groups, return an empty list
    if total_empty < separators:
        return []

    # Calculate the number of empty cells that can be distributed freely
    free_empty_cells = total_empty - separators

    # Create the base representation with groups and separators
    base_representation = []
    for group in groups:
        base_representation.append('F' * group)
    # Add the minimum required separators as single 'E's
    base_representation += ['E'] * separators
    # Add the remaining free empty cells as individual 'E's
    base_representation += ['E'] * free_empty_cells

    # Generate all permutations of the base representation
    all_permutations = set(permutations(base_representation))

    # Filter permutations to ensure valid arrangements
    valid_arrangements = []
    for perm in all_permutations:
        # Join the permutation into a single arrangement string
        arrangement = ''.join(perm)

        # Validate that every filled group is separated by at least one empty cell
        if is_valid_arrangement(arrangement, groups):
            valid_arrangements.append(arrangement)

    # Return all unique valid arrangements
    return valid_arrangements


def is_valid_arrangement(arrangement, groups):
    """
    Validate that every filled group in the arrangement is separated by at least one empty cell.
    :param arrangement: The arrangement string (e.g., 'FFFFEFFFFFFF')
    :param groups: List of group sizes (e.g., [4, 7])
    :return: True if valid, False otherwise
    """
    # Split the arrangement into chunks of 'F' and 'E'
    chunks = arrangement.split('E')

    # Remove empty chunks caused by consecutive 'E's
    chunks = [chunk for chunk in chunks if chunk]

    # Check that each group size matches the expected groups
    group_index = 0
    for chunk in chunks:
        if 'F' in chunk:
            # Check if this chunk matches the current group size
            if group_index >= len(groups) or len(chunk) != groups[group_index]:
                return False
            group_index += 1
        elif 'E' in chunk:
            # Skip empty cell chunks
            continue

    # Ensure all groups are accounted for
    return group_index == len(groups)


# Example usage
if __name__ == "__main__":
    results = generate_arrangements(20, [4, 7])
    print(f"Number of arrangements: {len(results)}")
    for arrangement in results:
        print(arrangement)
```
