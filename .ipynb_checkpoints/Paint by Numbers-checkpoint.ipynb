```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def shrink(img, factor=4):
    """
    Reduce resolution by specified factor (i.e. # pixels reduced by factor**2)
    """
    return np.round(np.mean(np.stack([img[np.expand_dims(np.arange(0, img.shape[0] - (img.shape[0] % factor), factor), 1) + r, 
                                          np.expand_dims(np.arange(0, img.shape[1] - (img.shape[1] % factor), factor), 0) + c]
                                      for r in range(factor)
                                      for c in range(factor)]), axis=0)).astype(np.uint8)

def get_prime_factors(n):
    """
    Decompose input integer into list of prime factors
    """
    prf = []
    pr = [2]
    while pr[-1] <= (n ** 0.5):
        if n % pr[-1] == 0:
            prf += [pr[-1]]
            n = int(n / pr[-1])
        else:
            pr += [pr[-1]+1]
            while any([pr[-1] % i == 0 
                       for i in pr[:-1]
                       if i <= (pr[-1] ** 0.5)]):
                pr[-1] += 1
    return [*prf, n]
    
def squarest_2d(fct):
    """
    Given an input integer representing a number of cells, pixels, etc., 
    identifies the squarest possible 2-d arrangement of the same number.
    E.g. 100 => (10, 10), 104 => (8, 13), 107 => (107, 1).
    Returns (1) the 2 dimension sizes, as list of ints, (2) "squareness" score.
    """
    if not isinstance(fct, list):
        fct = get_prime_factors(fct)
    if len(fct) == 1:
        return [*fct, 1], (1 / max(fct))
    elif len(fct) == 2:
        return fct, (min(fct) / max(fct))
    else:
        cposs = [sorted([int(j) for j in i.split(',')])
                 for i in {','.join([str(j) for j in i]) 
                           for i in [[*[fct[i] for i in range(len(fct)-1) if i not in [x, y]],
                                      fct[x]*fct[y], fct[-1]]
                                     for x in range(len(fct)-2) 
                                     for y in range(x+1, len(fct)-1)]}]
        c, sc = zip(*[squarest_2d(i) for i in cposs])
        bestix = [i for i in range(len(c)) if sc[i] == max(sc)][0]
        
        return c[bestix], sc[bestix]

def make_squarer(img):
    """
    Returns the input image array, with dimensions reshaped to have a "squareness" score 
    of 0.8 or more (per squarest_2d function).  Additional white pixels are added where 
    necessary to make this possible.
    """
    img = img.reshape(-1, 3)
    gsize = img.shape[0]
    while True:
        c, sc = squarest_2d(gsize)
        if sc > 0.8:
            break
        gsize += 1

    buff = np.repeat([[255]*3], gsize-img.shape[0], axis=0)

    return np.append(img, buff, axis=0).reshape(tuple([*c, 3]))

def img_to_palette(img, palette):
    """
    Convert image to palette colors using Euclidean Distance on RGB scores
    """
    img_pcode_rgb = np.zeros(img.shape[:2], dtype=int)
    mindiff = np.sum((img - np.expand_dims(palette[0], (0, 1))) ** 2, axis=2)
    for c in range(1, palette.shape[0]):
        diff = np.sum((img - np.expand_dims(palette[c], (0, 1))) ** 2, axis=2)
        img_pcode_rgb = np.where(diff < mindiff, c, img_pcode_rgb)
        mindiff = np.minimum(mindiff, diff)
    return img_pcode_rgb

def augment_palette(img, img_p, tol_sqerr=3000):
    """
    Add one mixed color to palette to reduce "poorly-matched" pixels in image.
    """
    # Count "poorly-matched" pixels; return empty array if < 1% of img
    badmatch_filt = np.sum((img_p - img) ** 2, axis=2).reshape(-1) > tol_sqerr
    if np.mean(badmatch_filt) < 0.01:
        return palette[:0], palette_mix[:0]

    # Identify poorly-matched colors, count frequencies
    badmatch_color = img.reshape(-1, 3)[badmatch_filt]
    badmatch_color = badmatch_color[np.lexsort([badmatch_color[:, 2-i] for i in range(3)])]
    fvix = np.arange(badmatch_color.shape[0]+1)\
               [np.any(np.append(badmatch_color, -np.ones_like(badmatch_color[:1], dtype=int), axis=0) 
                       != np.append(-np.ones_like(badmatch_color[:1], dtype=int), badmatch_color, axis=0),
                       axis=1)]
    badmatch_colors = badmatch_color[fvix[:-1]]
    badmatch_freq = fvix[1:] - fvix[:-1]

    # Get best mixed color match for each
    badmatch_altcode = np.zeros(badmatch_colors.shape[0], dtype=int)
    mindiff = np.sum((badmatch_colors - np.expand_dims(mixed_palette[0], 0)) ** 2, axis=1)
    for c in range(1, mixed_palette.shape[0]):
        diff = np.sum((badmatch_colors - np.expand_dims(mixed_palette[c], 0)) ** 2, axis=1)
        badmatch_altcode = np.where(diff < mindiff, c, badmatch_altcode)
        mindiff = np.minimum(mindiff, diff)

    # Add most frequently substituted mixed color to palette
    poss_mix = np.unique(badmatch_altcode)
    add_ix = poss_mix[np.argmax([np.sum(badmatch_freq[badmatch_altcode == c]) for c in poss_mix])]

    return mixed_palette[add_ix: add_ix+1], ', '.join(mixed_palette_ix[add_ix].astype(str))
```

```python
# Get palette
pbn = mpimg.imread('data/pbn.jpeg')
palette = np.stack([np.round(np.mean(pbn[r:r+5, c:c+10], axis=(0, 1))).astype(int)
                        for r in np.round(np.linspace(48, 528, 10)).astype(int)
                        for c in np.round(np.linspace(30, 315, 5)).astype(int)])
palette = np.delete(palette, [5, 15, 45, 49], axis=0)
palette_mix = np.array(['None']*palette.shape[0])
plt.imshow(make_squarer(palette))
```

```python
# Get possible palette from mixing color pairs
ix = np.column_stack((np.repeat(np.arange(palette.shape[0]), palette.shape[0]),
                      np.tile(np.arange(palette.shape[0]), palette.shape[0])))
ix = ix[ix[:, 1] > ix[:, 0]]
mixed_palette = ((palette[ix[:, 0]] + palette[ix[:, 1]]) / 2).astype(int)
mixed_palette_ix = ix
```

```python
# Get image
img = mpimg.imread('data/luna.jpg')
img = np.flip(np.transpose(img, (1, 0, 2)), axis=1)[216:-216, :2700]
img = shrink(img)
plt.imshow(img)
```

```python
# Convert img array to palette colors
while True:
    img_p_ix = img_to_palette(img, palette)
    img_p = palette[img_p_ix]
    palette_add, palette_mix_add = augment_palette(img, img_p, tol_sqerr=3000)
    if palette_add.shape[0]:
        palette = np.append(palette, palette_add, axis=0)
        palette_mix = np.append(palette_mix, palette_mix_add)
    else:
        break

plt.imshow(img_p)
```

```python
# Set up indices for small circles
circ_diam = 11
circ_ctr = (circ_diam-1)/2
circ_ix = np.array([[r, c]
                    for r in range(circ_diam)
                    for c in range(circ_diam)
                    if ((r-circ_ctr) ** 2) + ((c-circ_ctr) ** 2) < ((circ_diam/2) ** 2)])
circ_pixel_wts = (1 - (np.sqrt(np.sum((circ_ix - circ_ctr) ** 2, axis=1)) / (circ_diam/2))) ** 2
```

```python
plt.scatter((circ_diam/2) - np.sqrt(np.sum((circ_ix - circ_ctr) ** 2, axis=1)), 
            circ_pixel_wts, 
            s=2)
plt.show()
```

```python
# Flag most frequent color in each small circle
img_circ_p_ix = np.stack([img_p_ix[r:r+img_p_ix.shape[0]-circ_diam+1, 
                                   c:c+img_p_ix.shape[1]-circ_diam+1]
                          for r, c in circ_ix],
                         axis=-1).reshape(-1, circ_ix.shape[0])
img_circ_posix = np.column_stack((np.repeat(np.arange(img_p_ix.shape[0]-circ_diam+1), 
                                            img_p_ix.shape[1]-circ_diam+1),
                                  np.tile(np.arange(img_p_ix.shape[1]-circ_diam+1), 
                                          img_p_ix.shape[0]-circ_diam+1)))

img_circ_top_p_ix = -np.ones(img_circ_p_ix.shape[0], dtype=int)
img_circ_top_p_ct = np.zeros(img_circ_p_ix.shape[0], dtype=int)
for c in range(circ_ix.shape[1]):
    filt = np.all(img_circ_p_ix[:, c:c+1] != img_circ_p_ix[:, :c], axis=1)
    ct = np.sum(img_circ_p_ix[filt, c:] == img_circ_p_ix[filt, c:c+1], axis=1)
    img_circ_top_p_ix[filt] = np.where(ct > img_circ_top_p_ct[filt], 
                                       img_circ_p_ix[filt, c], 
                                       img_circ_top_p_ix[filt])
    img_circ_top_p_ct[filt] = np.where(ct > img_circ_top_p_ct[filt], 
                                       ct, 
                                       img_circ_top_p_ct[filt])
```

```python
# Use circle colors and pixel weights to blend pixel colors
img_p_ix_per_circ = -np.ones([*img_p_ix.shape, circ_ix.shape[0]], dtype=int)
img_p_ix_per_circ_wts = np.zeros([*img_p_ix.shape, circ_ix.shape[0]], dtype=float)
for i in range(circ_ix.shape[0]):
    img_p_ix_per_circ[img_circ_posix[:, 0] + circ_ix[i, 0],
                      img_circ_posix[:, 1] + circ_ix[i, 1],
                      i] = img_circ_top_p_ix
    img_p_ix_per_circ_wts[img_circ_posix[:, 0] + circ_ix[i, 0],
                          img_circ_posix[:, 1] + circ_ix[i, 1],
                          i] = circ_pixel_wts[i]
img_p_ix_per_circ[:, :, 0] = np.where(np.sum(img_p_ix_per_circ_wts != 0, axis=2) == 0, 
                                      img_p_ix, 
                                      img_p_ix_per_circ[:, :, 0])
img_p_ix_per_circ_wts[:, :, 0] = np.where(np.sum(img_p_ix_per_circ_wts != 0, axis=2) == 0, 
                                          1,
                                          img_p_ix_per_circ_wts[:, :, 0])

img_p_ix_per_circ = img_p_ix_per_circ.reshape(-1, circ_ix.shape[0])
img_p_ix_per_circ_wts = img_p_ix_per_circ_wts.reshape(-1, circ_ix.shape[0])
img_p_ix_per_circ_wts /= np.sum(img_p_ix_per_circ_wts, axis=1, keepdims=True)

pxl_top_p_ix = -np.ones(img_p_ix_per_circ.shape[0], dtype=int)
pxl_top_p_ct = np.zeros(img_p_ix_per_circ.shape[0], dtype=int)
for c in range(circ_ix.shape[1]):
    filt = (np.all(img_p_ix_per_circ[:, c:c+1] != img_p_ix_per_circ[:, :c], axis=1))\
           & (img_p_ix_per_circ[:, c] != -1)
    ct = np.sum(np.where(img_p_ix_per_circ[filt, c:] == img_p_ix_per_circ[filt, c:c+1],
                         img_p_ix_per_circ_wts[filt, c:c+1],
                         0.),
                axis=1)
    pxl_top_p_ix[filt] = np.where(ct > pxl_top_p_ct[filt], 
                                  img_p_ix_per_circ[filt, c], 
                                  pxl_top_p_ix[filt])
    pxl_top_p_ct[filt] = np.where(ct > pxl_top_p_ct[filt], 
                                  ct, 
                                  pxl_top_p_ct[filt])

img_p_ix_blended = pxl_top_p_ix.reshape(img_p_ix.shape)
img_p2 = palette[img_p_ix_blended]
```
