```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_line_pts(start, rlines):
    line_pts = [start]
    for rr, rc in rlines:
        line_pts += [(line_pts[-1][0]+rr, line_pts[-1][1]+rc)]
    
    return line_pts
```

```python
class ScaleDrawing(object):
    def __init__(self, height, width, margin=50):
        self.margin = margin
        self.pixels = np.zeros((height + (2 * self.margin), 
                                width + (2 * self.margin)), 
                               dtype=np.uint8)
        self.dims = self.pixels.shape

    def show(self):
        """
        Display drawing on screen
        """
        plt.figure(figsize=(12.8, 9.6))
        plt.imshow(self.pixels, cmap='gray_r', vmin=0, vmax=1) 

    def line(self, start, end, weight=1):
        """
        Draw line between specified start and end point
        """
        stx = np.argmax(np.abs([end[i]-start[i] for i in range(2)]))
        if end[stx] < start[stx]:
            end, start = (start, end)
        ofs = np.round(np.array([[end[0]-start[0], end[1]-start[1]]])).astype(int)
        nst = round(end[stx]) - round(start[stx]) + 1
        mag = np.sqrt(np.sum(ofs ** 2))
        
        bpts = (ofs * np.expand_dims(np.linspace(0, 1, nst), 1)) + weight
        adj = np.flip(ofs * [[1, -1]], axis=1) / mag * ((weight-1)/2)
        lim1 = np.round(bpts - adj).astype(int)
        lim2 = np.round(bpts + adj).astype(int)
        end1 = np.round(np.column_stack([np.linspace(lim1[0, i], lim2[0, i], 
                                                     lim1[0, stx]-lim2[0, stx]+1)
                                         for i in range(2)])).astype(int)
        end2 = np.round(np.column_stack([np.linspace(lim1[-1, i], lim2[-1, i], 
                                                     lim1[-1, stx]-lim2[-1, stx]+1)
                                         for i in range(2)])).astype(int)
        bpts = np.round(bpts).astype(int)
        
        crd1 = np.unique(np.append(lim1[:, stx], lim2[:, stx]))
        
        crd2_l = -np.ones(np.max(crd1)+1, dtype=int)
        crd2_l[bpts[:, stx]] = bpts[:, stx^1]
        crd2_h = -np.ones(np.max(crd1)+1, dtype=int)
        crd2_h[bpts[:, stx]] = bpts[:, stx^1]
        for x in [lim1, lim2, end1, end2]:
            crd2_l[x[:, stx]] = np.where(crd2_l[x[:, stx]] == -1, 
                                         x[:, stx^1], 
                                         np.minimum(crd2_l[x[:, stx]], x[:, stx^1]))
            crd2_h[x[:, stx]] = np.where(crd2_h[x[:, stx]] == -1, 
                                         x[:, stx^1], 
                                         np.maximum(crd2_h[x[:, stx]], x[:, stx^1]))
        
        npx = crd2_h[crd1]-crd2_l[crd1]+1
        crd2 = np.expand_dims(crd2_l[crd1], 1) + np.expand_dims(np.arange(np.max(npx)), 0)
        crd2 = np.where(crd2 > np.expand_dims(crd2_h[crd1], 1), -1, crd2).reshape(-1)
        crd2 = crd2[crd2 != -1]
        
        fillpts = np.column_stack((np.repeat(crd1, npx), crd2))
        if stx:
            fillpts = np.flip(fillpts, axis=1)
        fillpts = fillpts - weight + np.round([[*start]]).astype(int) + self.margin
        fillpts = fillpts[np.all(fillpts >= [[0, 0]], axis=1) 
                          & np.all(fillpts < [self.dims], axis=1)]
        
        fillpts = np.unique(np.sum(fillpts * [[self.dims[1], 1]], axis=1))
        
        self.pixels = self.pixels.reshape(-1)
        self.pixels[fillpts] = 1
        self.pixels = self.pixels.reshape(self.dims)

    def lines(self, pts, weight=1):
        """
        Draw lines between specified points.
        """
        for i in range(len(pts)-1):
            self.line(pts[i], pts[i+1], weight=weight)

    def symbol(self, pts, symbol_data):
        """
        Plot symbol at specified point(s)
        """
        r_os, c_os = [int((i-1)/2) for i in symbol_data.shape]
        for r, c in pts:
            self.pixels[r-r_os+self.margin:r-r_os+self.margin+symbol_data.shape[0], 
                        c-c_os+self.margin:c-c_os+self.margin+symbol_data.shape[1]] = symbol_data
```

```python
raw_lines = dict()
raw_lines['edge1'] = [(0, 713), (342.75, 0), (0, -285)]
raw_lines['edge2'] = [(994, 0), (0, 113.25)]

raw_fp = dict()
raw_fp['w_gaps'] = [23.5, 48, 35]
raw_fp['n_gaps'] = [96.75, 96.25, 96.25, 97.25, 94.75, 96.25, 97, 96.25, 96.5, 94.5, 32.25]
raw_fp['e_gaps_n'] = [41.75, 94.25, 96, 96.5, 67.25, 96.25, 37.75]
raw_fp['e_gaps_s'] = [54.5, 92.5]
raw_fp['s_gaps'] = [41.75, 94.25, 96.25, 95.25, 9.25]
```

```python
# Current
sd = ScaleDrawing(1000, 720)

sd.lines(get_line_pts((0, 0), raw_lines['edge1']))
sd.lines(get_line_pts((0, 0), raw_lines['edge2']))

fp = np.ones((9, 9), dtype=int)
fp[1:-1, 1:-1] = 0

sd.symbol([(0, round(i)) for i in np.cumsum(raw_fp['e_gaps_n'])], fp)
sd.symbol([(round(sum(raw_fp['n_gaps']) - i), 0) for i in [0, *np.cumsum(raw_fp['n_gaps'])[:-1]]], fp)
sd.symbol([(round(sum(raw_fp['n_gaps'])), int(i)) for i in np.cumsum(raw_fp['w_gaps'])], fp)
sd.symbol([(round(i), 713) for i in np.cumsum(raw_fp['s_gaps'])[:-1]], fp)

sd.symbol([(42, round(713-i)) for i in np.cumsum(raw_fp['e_gaps_s'])], fp)

sd.show()
```
