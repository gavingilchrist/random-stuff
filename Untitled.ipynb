```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_per_iter = 1000
n_iter = 1000000

est_n = 0
est_iter = []
for i in range(n_iter):
    est_n += int(np.sum(np.sum(np.random.random((n_per_iter, 2)) ** 2, axis=1) < 1))
    est_iter += [est_n * 4 / (i+1) / n_per_iter]

print(est_iter[-1])
plt.plot(np.arange(n_iter)+1, est_iter)
plt.plot(np.arange(n_iter)+1, np.repeat(np.pi, n_iter))
plt.ylim(3.13, 3.15)
plt.show()
```
