```python
def get_prime_factors(n):
    """
    Decompose input integer into list of prime factors
    """
    prf = []
    pr = [2]
    while pr[-1] <= (n ** 0.5):
        if n % pr[-1] == 0:
            prf += [pr[-1]]
            n = int(n / pr[-1])
        else:
            pr += [pr[-1]+1]
            while any([pr[-1] % i == 0 
                       for i in pr[:-1]
                       if i <= (pr[-1] ** 0.5)]):
                pr[-1] += 1
    return [*prf, n]
```

```python
def add_next_prime(d):
    p = d[-1]+1
    while any([p%i==0 for i in d]): p+=1
    return d+[p]

get_prime_factors = lambda n, d=None: ([n] if n < (d:=d or [2])[-1]**2 
                                       else get_prime_factors(n, add_next_prime(d)) if n % d[-1] 
                                       else [d[-1], *get_prime_factors(n//d[-1], d)])
```
